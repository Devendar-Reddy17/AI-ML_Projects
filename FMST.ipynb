Encoding above categorical data using get dummies and drop first

```python
credit_data_dummys = pd.get_dummies(credit_data, columns = ['account_check_status', 'credit_history', 'purpose', 'savings',
       'present_emp_since', 'personal_status_sex', 'other_debtors', 'property',
       'other_installment_plans', 'housing', 'job', 'telephone',
       'foreign_worker'])
```

```python
credit_data_dummys.shape
```

 Checking correlation of dummy data

```python
cor = credit_data_dummys.corr()
cor.loc[:,:] = np.tril(cor, k=-1)
cor=cor.stack()
cor[(cor > 0.55) | (cor < -0.55)]
```

```python
credit_data_dummys.columns
```

 Dropping irrelevant columns

```python
credit_data_dummys = credit_data_dummys.drop( ["foreign_worker_yes", "telephone_yes, registered under the customers name ", "job_unskilled - resident", "housing_rent", "other_installment_plans_none", "other_debtors_none",
                                             "personal_status_sex_male : single", "savings_unknown/ no savings account", "other_debtors_co-applicant"], axis=1 )
```

Separating target column and Splitting data

```python
from sklearn.model_selection import train_test_split

y = credit_data_dummys['default']
x = credit_data_dummys.loc[:, credit_data_dummys.columns != 'default']
x_train, x_test, y_train, y_test = train_test_split( x, y, test_size=0.3, random_state=42 )
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

applying random forest model

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier( n_jobs=2, n_estimators=500, criterion="entropy", random_state=9999 )
rfm = rf.fit(x_train, y_train)
```

```python
preds = rf.predict_proba(x_test)[:,1]
y_pred=rf.predict(x_test)
```

caluculating accuracy and confusion matrix

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
def cal_conf_matrix(y_true, y_pred):
    cm=confusion_matrix(y_true, y_pred)
    print(cm)
```

```python
cal_conf_matrix(y_test, y_pred)
print(accuracy_score(y_test, y_pred))
```

Checking feature importances

```python
importances = rf.feature_importances_
indices = np.argsort(importances)[::-1][:15]
a=credit_data_dummys.columns[:]
features = a.drop('default', 1)
plt.figure(1)
plt.title("Feature importances")
plt.barh(range(len(indices)), importances[indices], color='b', align = 'center')
plt.yticks(range(len(indices)), features[indices])
plt.xlabel("Relative Importance")
```

```python
trainResult = rf.score(x_train, y_train)
testResult = rf.score(x_test, y_test)
```

```python
"Train Accuracy:"
(trainResult*100.0)
```

```python
"Test Accuracy:"
(testResult*100.0)
```
